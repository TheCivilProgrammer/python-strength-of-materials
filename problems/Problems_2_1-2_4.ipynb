```python
import numpy as np
from sympy import symbols, Eq, solve
import sympy

def radians(degrees):
  return degrees * np.pi/180


class Force:
  def __init__(self, name = "", magnitude = None, direction = None):
    self.magnitude = symbols(name + '_' + 'magnitude')
    if magnitude is not None: 
      self.magnitude = magnitude

    # direction is in degrees counterclockwise from axis x
    self.direction = symbols(name + '_' + 'direction')
    if direction is not None: 
      self.direction = direction

    self.x, self.y = self.decompose()

  def decompose(self):
    x = self.magnitude * sympy.cos(radians(self.direction))
    y = self.magnitude * sympy.sin(radians(self.direction))
    return x, y


def get_resultant(list_of_forces):
  Rx = np.sum([force.x for force in list_of_forces])
  Ry = np.sum([force.y for force in list_of_forces])
  R = sympy.sqrt(sympy.square(Rx)+sympy.square(Ry))
  tan_alpha = Ry/Rx
  alpha = sympy.degrees(sympy.arctan(tan_alpha))
  return R, alpha
```

```python
X1 = Force(magnitude= 900, direction= 30)
X2 = Force(magnitude= 600, direction= 75)
list_of_forces = [X1, X2]
R = get_resultant(list_of_forces)

print(R)
```
