```python
from sympy import symbols, Eq, solve
import numpy as np
```

```python
P = symbols('P')

eq = 425 * np.cos(np.radians(30)) - P * np.cos(np.radians(20))
solve(eq)
```

```python
class Force:
  def __init__(self, magnitude, direction):
    self.magnitude = magnitude
    # direction is in degrees counterclockwise from axis x
    self.direction = direction
    self.x, self.y = self.decompose()

  def decompose(self):
    x = self.magnitude * np.cos(np.radians(self.direction))
    y = self.magnitude * np.sin(np.radians(self.direction))
    return x, y

def get_resultant(list_of_forces):
  Rx = np.sum([force.x for force in list_of_forces])
  Ry = np.sum([force.y for force in list_of_forces])
  R = np.sqrt(np.square(Rx)+np.square(Ry))
  if abs(Rx) < 0.01:
    if Ry >= 0:
      alpha = 90
    else: 
      alpha = -90
  else:  
    tan_alpha = Ry/Rx
    alpha = np.degrees(np.arctan(tan_alpha))
  return R, alpha
```

```python
Q = Force(magnitude = 425, direction = 180-30)
P = Force(magnitude = 391.682118670422, direction = 20)

list_of_forces = [Q, P]
get_resultant(list_of_forces)
```
