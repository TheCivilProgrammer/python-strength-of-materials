```python
import numpy as np
from sympy import symbols, Eq, solve
```

```python
class Force:
  def __init__(self, name = "", magnitude = None, direction = None):
    self.magnitude = symbols(name + '_' + 'magnitude')
    if magnitude is not None: 
      self.magnitude = magnitude

    # direction is in degrees counterclockwise from axis x
    self.direction = symbols(name + '_' + 'direction')
    if direction is not None: 
      self.direction = direction

    self.x, self.y = self.decompose()

  def decompose(self):
    x = self.magnitude * np.cos(np.radians(self.direction))
    y = self.magnitude * np.sin(np.radians(self.direction))
    return x, y

```

```python
T1 = Force(name='T1', direction=30)
T2 = Force(name='T2', direction=-45)
```

T1.x + T2.x - 5000 = 0  

T1.y + T2.y - 0 = 0


```python
eq1 = Eq(T1.x + T2.x - 5000)
eq2 = Eq(T1.y + T2.y - 0)

solve((eq1,eq2), (T1.magnitude, T2.magnitude))
```

# b)

```python
alpha_values = range(-90, 0)
T2_values = []

for alpha in alpha_values:
  T1 = Force(name='T1', direction=30)
  T2 = Force(name='T2', direction=alpha)

  eq1 = Eq(T1.x + T2.x - 5000)
  eq2 = Eq(T1.y + T2.y - 0)
  solution = solve((eq1,eq2), (T1.magnitude, T2.magnitude))

  T2_values.append(solution[T2.magnitude])
```

```python
index_of_min_T2_value = np.argmin(T2_values)

alpha_values[index_of_min_T2_value]
```
